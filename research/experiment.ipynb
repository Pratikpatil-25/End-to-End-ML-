```python
x = wine.drop(['quality'], axis = 1)
y = wine['quality']
```

```python
x.head()
```

```python
y.head(6)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(f"x_train:{x_train.shape}")
print(f"x_test:{x_test.shape}")
print(f"y_train:{y_train.shape}")
print(f"y_test:{y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression, ElasticNet
```

```python
el = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
el.fit(x_train, y_train)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
```

```python
def eval(actual, pred):
    rmse = root_mean_squared_error(actual, pred)
    r2 = r2_score(actual, pred)
    mae = mean_absolute_error(actual, pred)

    return rmse, r2, mae
```

```python
el_pred = el.predict(x_test)
lr_pred = lr.predict(x_test)
```

```python
el_rmse, el_r2, el_mae = eval(y_test, el_pred)
lr_rmse, lr_r2, lr_mae = eval(y_test,lr_pred)
```

```python
print(f"RMSE : \n LinearRegression : {lr_rmse} \n ElasticNet : {el_rmse}")
print(f"R2Score : \n LinearRegression : {lr_r2} \n ElasticNet : {el_r2}")
print(f"MAE : \n LinearRegression : {lr_mae} \n ElasticNet : {el_mae}")
```
